Import libraries:

```python
import pandas as pd
import pickle
```

```python
DATA_PATH = './data/'

ARTICLES_DATA = DATA_PATH+'articles.tsv.gz'
TOPICS_DATA = DATA_PATH+'topics.jsonl.gz'
MAPPING_DATA = DATA_PATH+'mapping.pickle'

COMPRESSION = 'gzip'

```

```python
# Read datasets
df_articles = pd.read_csv(ARTICLES_DATA, compression=COMPRESSION,sep='\t', error_bad_lines=False) #error_bad_lines --> drop lines with too many fields
df_topics = pd.read_json(TOPICS_DATA, compression=COMPRESSION, lines=True)
mapping_data = pickle.load(open(MAPPING_DATA, "rb"))
```
